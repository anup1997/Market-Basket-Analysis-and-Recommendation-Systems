```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("a.csv",delimiter=',',header=None)
df.columns = ['Items']
df['Trans'] = np.arange(1,len(df)+1)
df.head(30)
```

```python
cols = df.columns.tolist()
cols = cols[::-1]
df = df[cols]
df.head(30)
```

```python
df['Item-List'] = df["Items"].apply(lambda x: x.split(','))
df.head(30)
```

```python
item_dict = {}
for item in df['Item-List']:
    for i in item:
        if i not in item_dict.keys():
            item_dict[i] = 1
        else:
            item_dict[i] += 1
print(item_dict)
```

```python
import pandas as pd

number = [5,13,7,6,7,4,2,5,8,3,6]
items = ["MILK","Bread","Biscuit","Cornflakes","Tea","Bournvita","Jam","Maggi","Coffee","Cock","Suger"]

df = pd.DataFrame(list(zip(number, items)),
                  columns=['Quantity', 'Item'])
df.head()
```

```python
y = pd.get_dummies(df.Item, prefix='Items')
print(y.head())
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
y = LabelBinarizer().fit_transform(df.Item)
y
```

```python
hot_encoded_df = df.groupby(['Quantity','Item'])['Item'].count().unstack().reset_index().fillna(0).set_index('Quantity')
hot_encoded_df.head(10)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    else:
        return 1
hot_encoded_df = hot_encoded_df.applymap(encode_units)
```

```python
hot_encoded_df.head(10)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
femsreq_itets = apriori(hot_encoded_df, min_support=0.1,use_colnames = True)
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, metric='lift')
rules.head(10)
```

```python
rules[ (rules['lift'] > 1.1) & (rules['confidence'] > 0.4)]
```
